```python
std_dev = df['Unit Price'].std()
```

```python
mean = df['Unit Price'].mean()
```

```python
cutoff = std_dev * 3
```

```python
lower_cutoff = mean - cutoff
upper_cutoff = mean + cutoff
```

```python
cleaned_data = df[df['Unit Price'] > lower_cutoff]
cleaned_data = df[df['Unit Price'] < upper_cutoff]
```

```python
Q1 = np.percentile(cleaned_data['Unit Price'],25,interpolation='midpoint')
Q2 = np.percentile(cleaned_data['Unit Price'],50,interpolation='midpoint')
Q3 = np.percentile(cleaned_data['Unit Price'],75,interpolation='midpoint')
IQR = Q3 - Q1
```

```python
upper_cutoff = Q3 + 1.5*IQR
lower_cutoff = Q1 - 1.5*IQR
```

```python
cleaned_data = cleaned_data[cleaned_data['Unit Price'] > lower_cutoff]
cleaned_data = cleaned_data[cleaned_data['Unit Price'] < upper_cutoff] 
```
